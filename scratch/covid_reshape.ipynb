```python
import pandas

df = pandas.read_csv('raw_data/extended_covid.csv')
df

    
```

```python
temp = df['cases'][0]
for index, row in df.iterrows():
    if index == 0: continue
    cases = df['cases'][index] 
    
    if cases > temp:
        df['cases'][index] = cases - temp
    else:
        df['cases'][index] = temp - cases 
    temp = cases
    
```
